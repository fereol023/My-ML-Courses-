```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import *
```

```python
np.random.seed(23)

y = np.array([4,5,1])
y_pred = np.array([0,6,11])

print("MAbsoluteError : ", mean_absolute_error(y, y_pred))
print("MSE : ", mean_squared_error(y, y_pred)) # utile pour les cas où une erreur même petite serait grave
print("RMSE : ", np.sqrt(mean_absolute_error(y, y_pred)))
print("MedianAE : ", median_absolute_error(y, y_pred)) # est l'erreur entre 2 et 0
print("R2 : ", r2_score(y, y_pred))

err_hist = np.abs(y - y_pred)
plt.hist(err_hist) 
plt.show()
plt.scatter(np.arange(3), y)
plt.scatter(np.arange(3), y_pred)
plt.show()
```

```python
np.random.seed(23)

y = np.random.normal(0,1,500)
y_pred = np.random.normal(0,1,500)+np.random.normal(1.5)

print("MAbsoluteError : ", mean_absolute_error(y, y_pred))
print("MSE : ", mean_squared_error(y, y_pred))
print("RMSE : ", np.sqrt(mean_absolute_error(y, y_pred)))
print("MedianAE : ", median_absolute_error(y, y_pred))
print("R2 : ", r2_score(y, y_pred))

err_hist = np.abs(y - y_pred)
plt.hist(err_hist, bins=50) # nb de classes
plt.show()
```

```python
x = np.arange(500)
plt.scatter(x, y)
plt.scatter(x, y_pred)
```

Le R2 est 1- (SCResidus/SCT) = SCEstimes/SCT. Donne la part de la variabilité des données (ie de la variance) expliquée par le modèle; C'est aussi le model.score(X,y) après une regression

S'interprete aussi comme une mesure de ratio vraiance des erreurs/variance des données. erreurs << variance alors R2 proche de 1
